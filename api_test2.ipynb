```python
chart_df=requests.get(chart).text
print(chart_df)
```

```python
dataframe = pd.read_json(chart_df)
dataframe
```

```python
df = dataframe.drop(['volume','change','changeOverTime','changePercent','label','unadjustedVolume','vwap'],axis=1)
df
```

```python
df.head()

data = []

for col in df.columns:
    data.append(go.Candlestick(x= df.date, 
                       open=df.open,
                       high=df.high,
                       low=df.low,
                       close=df.close) )

url = py.iplot(data, filename=f'{stock}candlestick chart')
url
```
